```
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from scipy.interpolate import BSpline, make_interp_spline
```

```
x = np.array([0.5, 1.5, 2.5, 4, 5.5, 6.5, 7.5, 9, 10.5, 11.5, 12.5, 14, 15.5, 16.5, 17.5])
p = np.array([0, -32, 0, 72, 0, -32, 0, 72, 0, -32, 0, 72, 0, -32, 0])
v = np.array([3.5, 0.4, 0, -0.6, -3.5, -0.4, 0, 0.6, 3.5, 0.4, 0, -0.6, -3.5, -0.4, 0])
i = np.array([0, -80, -3.5, -120, 0, 80, 3.5, 120, 0, -80, -3.5, -120, 0, 80, 3.5])

x_new = np.linspace(0,17.5 , 300)
a_BSpline = sp.interpolate.make_interp_spline(x, p)


p_new = a_BSpline(x_new)



plt.plot(x, p, 'o')

plt.plot(x_new, p_new)
plt.legend(["Data", "p Curve"], loc='upper right', fontsize='x-small')

plt.xlabel("t (us)")
plt.ylabel("p (mW)")
plt.grid(b=True, which='major', color='#666666', linestyle='-', alpha=0.5)
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)


              
plt.show()


```
